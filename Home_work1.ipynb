## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
def rotate90(a, direction):
    if direction == "right":
        return [list(row)[::-1] for row in zip(*a)]
    elif direction == "left":
        return [list(row) for row in zip(*a[::-1])]
    else:
        raise ValueError("нужно написать либо left либо right")
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
rotate90(matrix, "right")
````

## Problem 3

Write a function that takes a string as input and returns a dictionary containing the number of occurrences of each character in the string. Use method `count` of strings. Implement this function in one line using $\lambda$-function.

Example for the string 'hello, world!': {'h': 1, 'e': 1, 'l': 3, 'o': 2, ',': 1, ' ': 1, 'w': 1, 'r': 1, 'd': 1, '!': 1}.

````python
count_chars = lambda s: {c: s.count(c) for c in s}
count_chars("hello, world!")
````

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
class Book:
    def __init__(self, title, author, year_publication, isbn, copies):
        self.title = title
        self.author = author
        self.year_publication  =  year_publication
        self.isbn = isbn
        self.copies = copies
        print(f"Book {title} created")
    def get_info(self):
        return f"Title: {self.title} Author:{self.author} Year:{self.year_publication} ISBN:{self.isbn} copies:{self.copies}"
    def issue_book(self):
        if self.copies > 0:
            self.copies -=1
            print(f"{self.title} issued")
            return True
        else:
            print(f"Sorry, {self.title} is out of stock.")
            return False
    def return_book(self):
        self.copies+=1
        print(f"{self.title} returned")
        return True
class User:
    def __init__(self, user_name, card_number):
        self.user_name = user_name
        self.card_number = card_number
        self.borrowed_books = []
        print(f"User {user_name} created")
    def add_book(self, book):
        if book not in self.borrowed_books:
            self.borrowed_books.append(book)
            return True
        else:
            print(f"Book already borrowed")
            return False
    def remove_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            return True
        else:
            print(f"Book not found in borrowed list")
            return False
class Library:
    def __init__(self):
        self.transaction_history = []
        self.list_of_books = []
        self.list_of_users = []
        print("Library created")
    def add_book(self, book):
        if book not in self.list_of_books:
            self.list_of_books.append(book)
            return True
        else:
            print(f"Book already in Library")
            return False
    def remove_book(self, book):
        if book in self.list_of_books:
            self.list_of_books.remove(book)
            return True
        else:
            print(f"Book not found in Library")
            return False       
    def add_user(self, user):
        if user not in self.list_of_users:
            self.list_of_users.append(user)
            return True
        else:
            print(f"User already in list")
            return False
    def remove_user(self,user):
        if user in self.list_of_users:
            self.list_of_users.remove(user)
            return True
        else:
            print(f"User not found in library")
            return False
    def issue_book(self, user, book):
        if user not in self.list_of_users:
            return False
        if book not in self.list_of_books:
            return False
        if book in user.borrowed_books:
            return False
        if not book.issue_book():
            return False
        if not user.add_book(book):
            book.return_book()
            return False
        self.transaction_history.append(f"Name: {user.user_name}, Book:{book.title} ISSUED")
        print(f"Book is given")
        return True
    def return_book(self, user, book):
        if user not in self.list_of_users:
            return False
        if book not in self.list_of_books:
            return False
        if book not in user.borrowed_books:
            return False
        user.remove_book(book)
        book.return_book()
        self.transaction_history.append(f"Name: {user.user_name}, Book:{book.title} RETURNED")
        return True

        
lib = Library()

b1 = Book("1984", "George Orwell", 1949, "ISBN-001", 1)
b2 = Book("Dune", "Frank Herbert", 1965, "ISBN-002", 2)

u1 = User("Alice", "C001")
u2 = User("Bob", "C002")

assert lib.add_book(b1) is True
assert lib.add_book(b2) is True
assert lib.add_book(b1) is False  

assert lib.add_user(u1) is True
assert lib.add_user(u2) is True
assert lib.add_user(u1) is False  


assert lib.issue_book(u1, b1) is True
assert b1.copies == 0
assert b1 in u1.borrowed_books
assert lib.transaction_history[-1].endswith("ISSUED")
````
